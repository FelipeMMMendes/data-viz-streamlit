```python
import pandas as pd
import streamlit as st
import plotly.express as px
```

```python
df1 = pd.read_csv('/content/Dataset_FireWatch_Brazil_Q1_2024.csv', sep=',',encoding='utf-8')
df2 = pd.read_csv('/content/Dataset_FireWatch_Brazil_Q2_2024.csv', sep=',',encoding='utf-8')
df3 = pd.read_csv('/content/Dataset_FireWatch_Brazil_Q3_2024.csv', sep=',',encoding='utf-8')
df4 = pd.read_csv('/content/Dataset_FireWatch_Brazil_Q4_2024.csv', sep=',',encoding='utf-8')

df = pd.concat([df1, df2, df3, df4])

df
```

```python
df['data'] = pd.to_datetime(df['data'])
df = df.sort_values('data')

df['data']
```

```python
df['Month'] =  df['data'].apply(lambda x: str(x.year) + '-' + str(x.month))
month = st.sidebar.selectbox('Mês', df['Month'].unique())

```

```python
df_filtered = df[df['Month'] == month]
df_filtered
```

```python
col1, col2 = st.columns(2)
col3, col4, col5 = st.columns(3)


queim_dia = px.bar(df_filtered, x='data', y='avg_risco_fogo', title='Risco de fogo Diário')
col1.plotly_chart(queim_dia)
```

```python
col1, col2 = st.columns(2)
col3, col4, col5 = st.columns(3)


queim_estado = px.bar(df_filtered, x='estado', y='avg_risco_fogo', title='Risco de fogo por Estado')
col1.plotly_chart(queim_estado)
```
